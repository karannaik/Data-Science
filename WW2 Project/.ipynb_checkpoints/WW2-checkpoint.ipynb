```python
# "Ctrl + Shift + p"  for shortcuts.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
print(os.listdir("input"))
```

```python
import warnings
warnings.filterwarnings("ignore")# if there is a warning after some codes, this will avoid us to see them.
```

```python
operationsFile = pd.read_csv("input/operations.csv")

operationsFile.head()
```

```python
weatherSummaryFile = pd.read_csv("input/Summary of Weather.csv")
weatherSummaryFile.head()
```

```python
weatherCountriesFile = pd.read_csv("input/Weather Station Locations.csv")
weatherCountriesFile.head()
```

```python
operationsFile = operationsFile[pd.isna(operationsFile["Country"])==False]
operationsFile = operationsFile[pd.isna(operationsFile["Target Latitude"])==False]
operationsFile = operationsFile[pd.isna(operationsFile["Target Longitude"])==False]
operationsFile.head()
```

```python
operationsFile.columns
```

```python
# drop unused features
drop_list = ['Mission ID','Unit ID','Target ID','Altitude (Hundreds of Feet)','Airborne Aircraft',
             'Attacking Aircraft', 'Bombing Aircraft', 'Aircraft Returned',
             'Aircraft Failed', 'Aircraft Damaged', 'Aircraft Lost',
             'High Explosives', 'High Explosives Type','Mission Type',
             'High Explosives Weight (Pounds)', 'High Explosives Weight (Tons)',
             'Incendiary Devices', 'Incendiary Devices Type',
             'Incendiary Devices Weight (Pounds)',
             'Incendiary Devices Weight (Tons)', 'Fragmentation Devices',
             'Fragmentation Devices Type', 'Fragmentation Devices Weight (Pounds)',
             'Fragmentation Devices Weight (Tons)', 'Total Weight (Pounds)',
             'Total Weight (Tons)', 'Time Over Target', 'Bomb Damage Assessment','Source ID']
operationsFile.drop(columns=drop_list, axis = 1, inplace=True)
operationsFile = operationsFile[ operationsFile.iloc[:,8]!="4248"] # drop this takeoff latitude 
operationsFile = operationsFile[ operationsFile.iloc[:,9]!=1355]   # drop this takeoff longitude
operationsFile.columns
```

```python
operationsFile.info()
```

```python
#Country counts:
print(operationsFile["Country"].value_counts())
plt.figure(figsize=(20,20))
sns.countplot(operationsFile["Country"])
plt.show()
```

```python
# Top target countries
topTargetCountries = operationsFile["Target Country"].value_counts()[:10]
print(topTargetCountries)
plt.figure(figsize=(20,20))
plt.xticks(rotation=90)
sns.countplot(operationsFile["Target Country"])
plt.show()
```

```python
# Aircraft Series
print(operationsFile["Aircraft Series"].value_counts()[:10])

plt.figure(figsize=(20,20))
sns.countplot(operationsFile["Aircraft Series"])
plt.xticks(rotation=90)
plt.show()
```
